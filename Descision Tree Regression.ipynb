## Import library yang dibutuhkan

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

## Load dataset yang akan digunakan

```python
data = pd.read_csv('mobil/audi_edit.csv')
data
```

## Bagi data menjadi variabel prediktor dan respon

Pada bagian ini data yang digunakan sebagai prediktor adalah mileage sementara data response adalah price

```python
X = data.iloc[:,4:5].values
y = data.iloc[:,2].values
```

## Proses fitting model regression

```python
# Fit regression model
model = DecisionTreeRegressor(random_state = 0)
model.fit(X, y)
```

## Visualiasasi tree dalam bentuk teks

```python
text_representation = tree.export_text(model)
print(text_representation)
```

## Visualisasi tree dengan plot

```python
fig = plt.figure(figsize=(125,100))
_ = tree.plot_tree(model, filled=True)
```

## Load data testing

```python
data_test = pd.read_csv('mobil/audi_edit2.csv')
X_test = data_test.iloc[:,4:5].values
y_test = data.iloc[:,2].values
```

## Proses prediksi dari data testing

```python
yhat = model.predict(X_test)
yhat
```

## Visualiasi hasil training descision tree

```python
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1)) 
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, model.predict(X_grid), color = 'blue')
plt.title('Mileage mobil & harga mobil')
plt.xlabel('Mileage Mobil')
plt.ylabel('Harga Mobil')
plt.show()
```

## Perhitungan keakuratan hasil regresi


```python
# Hasil MAE
mae = mean_absolute_error(y_test, yhat)
mse = mean_squared_error(y_test, yhat)
rmse = mean_squared_error(y_test, yhat, squared= False)

print("========= Hasil Evaluasi =========")
print('Hasil MAE: %.2f MAE' % (mae))
print('Hasil MSE: %.2f MSE' % (mse))
print('Hasil RMSE: %.2f RMSE' % (rmse))
```

## Komparasi data dan hasil prediksi regresi dalam bentuk grafik

```python
compare = pd.DataFrame(list(zip(y_test, yhat)), columns =['Real Value', 'Predict Value'])
compare.plot()
```
